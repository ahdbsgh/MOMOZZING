```python
train_data = all_data[:len(train)]
test_data = all_data[len(train):]
```

```python
train_data
```

```python
train_data = train_data.drop('Survived', 1)
```

```python
test_data = test_data.drop('Survived', 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_data, TARGET, test_size = 0.2,random_state=SEED)
```

# 모델링
