```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 계층적 군집분석 

```python

X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
import matplotlib.pyplot as plt

labels = range(1, 11)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(X, 'single')
```

```python

labelList = range(1, 11)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
data = customer_data.iloc[:, 3:5].values
```

## boston house 집값 데이터


```python
# 데이터 불러오기
data = pd.read_csv(r"C:\Users\home\WorkSpace\바이트디그리_머신러닝\Boston_house.csv") 

'''
타겟 데이터
1978 보스턴 주택 가격
506개 타운의 주택 가격 중앙값 (단위 1,000 달러)

특징 데이터
CRIM: 범죄율
INDUS: 비소매상업지역 면적 비율
NOX: 일산화질소 농도
RM: 주택당 방 수
LSTAT: 인구 중 하위 계층 비율
B: 인구 중 흑인 비율
PTRATIO: 학생/교사 비율
ZN: 25,000 평방피트를 초과 거주지역 비율
CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
AGE: 1940년 이전에 건축된 주택의 비율
RAD: 방사형 고속도로까지의 거리
DIS: 직업센터의 거리
TAX: 재산세율'''

data.head() # 데이터 확인
```

```python
target=data['Target']
data = data.drop(['Target'], axis = 1) 
```

```python
data.shape
```

```python

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
np.mean([x for x, y in zip(target,cluster.fit_predict(data)) if  y == 0])  #1번으로 예측한 타겟값들 #집값이 낮은 데이터들이 할당
```

```python
np.mean([x for x, y in zip(target,cluster.fit_predict(data)) if  y == 1]) #2번으로 예측한 타겟값들 #집값이 높은 데이터들이 할당
```
